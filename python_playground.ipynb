```python
# Python fundamental 
class Student:
    # class varible
    student_num = 0

    def __init__(self, name, sex):
        self.name = name
        self.sex = sex 
        Student.student_num += 1 

    # class method
    @classmethod
    def add_students(cls, add_num):
        cls.student_num += add_num

    # using the class method to replace the construction method 
    @classmethod
    def from_string(cls, info):
        name, sex = info.split(' ')
        return cls(name, sex)

    # can not access class attributes and instance 
    @staticmethod
    def name_len(name):
        return len(name)

s1 = Student('Qiqi','Female')
s2 = Student.from_string('Qangqang Male')
print(f'Student.student_num: ', Student.student_num)

# Calling the static method
# need to use the classname to call the static method
print(f's2.name: {s2.name}, s2.name.len: {Student.name_len(s2.name)}')
```

```python
# generator - 生成器

""" 
    1. memory efficiency - lazy evaluation, compute only when needed  
    2. minimum memory usage -> store state, not full data
    3. Fast execution -> resume instead of restarting 
    4. Pipeline friendly -> idea for streaming big or infinite data
"""


# 1. method as generator, 使用yield
# yield 类似于return，但不会结束method
def square_number(nums):
    result = []
    for i in nums:
        result.append(i*i)
    return result 

## 
def square_number(nums):
    for i in nums:
        yield i*i


my_gen = square_number([1,2,3,4,5])
# print the physical location of the generator value
print(my_gen)  
# how to print the generator values 
print('print next function:')
print(next(my_gen))

print('Then running for loop:')
for i in my_gen:
    print(i)
    
    
# method 2 to craete a generator, using the comprehension 
# notice: using ()
print('method 2 - to craete the generator')
my_gen2 = (i*i for i in [1,2,3,4,5])
for i in my_gen2:
    print(i)


# -- verfit the advantage of using generator 
import psutil, time, os, random 

random_nums = [i for i in range(10)]

list_start_time = time.time() 
list_start_memo = psutil.Process(os.getpid()).memory_info().rss/1024/1024

nums_list = [random.choice(random_nums) for _ in range(1000000)]

list_end_time = time.time() 
list_end_memo = psutil.Process(os.getpid()).memory_info().rss/1024/1024

print(f""" Before building list, program takes memory: {list_start_memo} MB
      Took time: {list_end_time - list_start_time} seconds
      After building generator, program takes memory: {list_end_memo} MB
      """)

gen_start_time = time.time() 
gen_start_memo = psutil.Process(os.getpid()).memory_info().rss/1024/1024

num_gen = (random.choice(random_nums) for _ in range(1000000))

gen_end_time = time.time() 
gen_end_memo = psutil.Process(os.getpid()).memory_info().rss/1024/1024

print(f""" Before building generator, program takes memory: {gen_start_memo} MB
      Took time: {gen_end_time - gen_start_time} seconds
      After building generator, program takes memory: {gen_end_memo} MB
      """)
```

```python
# Decorator,
# a method that accept another func as input, and then output the same func 
# function accept function as input arguments, 将函数传递给函数

def square(x): 
    return x*x

def print_running(f, x):
    print(f'{f.__name__} is running')
    return f(x)

result = print_running(square, 2)
print(result)

# example 2 
import time 

def decorator(func):
    def wrapper(*args, **kwargs):
        # *args, **kwargs --》 因为不知道传入的变量的数量，所以这两个可以cover all cases
        start_time = time.time()
        result = func(*args, **kwargs)
        end_time = time.time()
        print(f'{func.__name__} execution time: {start_time - end_time}')
        return result
    return wrapper

decorated_square = decorator(square)
decorated_square(10) 

@decorator
def square(x):
    return x * x 

square(10)


# example 3, decorator generator 
import functools

def timer(threshold):
    def decorator(func):
        def wrapper(*args, **kwargs):
            start_time = time.time()
            result = func(*args, **kwargs)
            end_time = time.time()
            if end_time - start_time > threshold:
                print(f'{func.__name__} took longer than {threshold}')
            return result
        return wrapper
    return decorator 

@timer(0.2)
def sleep_04():
    time.sleep(0.4)

sleep_04 = timer(0.2)(sleep_04)
sleep_04()


""" 
    1. 增加代码复用
    2. 保证代码逻辑清晰， 确保单一职责
    3. 用装饰器可以扩展别人的函数
"""
```

```python
# 魔法方法 - dunder method, dunder - double underscore 

# exmaple:
""" 
__add__()
__init__()

"""

class ShoppingCart:
    def __init__(self, items:list[str]):
        self.items = items 
        
    def __add__(self, another_cart):
        new_cart = ShoppingCart(self.items + another_cart.items)
        return new_cart
    
    def __str__(self):
        return f'Cart({self.items})'        
    
    def __len__(self):
        return len(self.items)
    
    def __call__(self, item):
        self.items.append(item)
        
            
cart1 = ShoppingCart(['Apple', 'Banana'])
cart2 = ShoppingCart(['Orange', 'Pear'])
new_cart = cart1 + cart2
print(new_cart.items)

cart = ShoppingCart(['Apple2', 'Banana2'])
print(len(cart))

```

```python
# Match Case 

# example:
def if_traffic_light(color):
    match color:
        case 'red':
            return 'Stop'
        case 'yellow':
            return 'Caution'
        case 'green':
            return 'Go'
        case _:
            return 'Invalid color'

""" 
    解包并且绑定常量
    不论带什么括号 或者不带括号
    case可以直接带数据类型 - case tuple:
    
"""
def match_case(point:tuple):
    match point:
        case (0,0):
            print('Origin') 
        case (x,y):
            print(f'{x=}, {y=}')
        case others:
            print(f'{others} is not a valid point')

p1 = (0,2)
p2 = (1,2,3)
match_case(p1)
match_case(p2)
        
```

```python
# dataclass 
""" 
    When: 当class主要是用来存醋数据
    想什么dataclass自动生成方法的特性
"""
# using calss 
class NormalPerson:
    def __init__(self, name, age):
        self.name = name 
        self.age = age

from dataclasses import dataclass, field 
import random
from typing import ClassVar

@dataclass(order=True)
class Person:
    name:str
    age:int
    height:int = field(default_factory=lambda: random.randint(150,200))
    people_num: ClassVar[int] = 0
    
    def __post_init__(self):
        Person.people_num += 1 
        
person1 = NormalPerson("Alice", 20)
person2 = Person("Bob", 20)

print(person1)
print(person2)
```
